```python
import Simfunc as sim
import matplotlib.pyplot as plt
import numpy as np
```

# 1 source

```python
#Define Mask
mask_len = 5000
mask = sim.def_mask(mask_len)
```

```python
#Open rate
open_cell = np.sum(mask == 1)
print(open_cell)
print("Mask's Open rate is {0}".format(open_cell/mask_len))
```

```python
#Define Source Position
ang = [30]
flux = [1]
pos = sim.def_source_pos(ang, flux)
plt.plot(pos, marker = '*', linestyle = "None")
plt.xlabel("zenith angle")
plt.ylabel("Source Flux")
#plt.savefig("1_source_source_pos.png")
```

```python
#Calc Detector image
detec = sim.create_detec_image(mask_len, mask, ang,flux, fl=mask_len)

plt.plot(detec)
plt.xlabel("detector pixel")
plt.ylabel("Counts")
plt.xlim(12000, 12500)
plt.ylim(0.8, 1.2)
#plt.savefig("./1_source_detec_imag.png")
```

```python
#Decode image
result = sim.decode_image(mask, detec, mask_len)

plt.plot(result[0], result[1])
plt.yscale("log")
plt.xlim(20, 40)
plt.ylim(8e2, 3e3)
plt.xlabel("zenith angle")
plt.ylabel("decoded source Flux")
#plt.savefig("1_source_decode_imag.png")
```
